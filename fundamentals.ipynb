Schreibe eine Funktion summarize_types(data)

→ nimmt eine Liste gemischter Datentypen und zählt, wie oft jeder Typ vorkommt (nutze type(x).__name__).

Beispiel: [1, 2.0, "hi", True, 3] → {'int': 2, 'float': 1, 'str': 1, 'bool': 1}


Bonus: Finde heraus, warum True + True == 2 in Python.

```python
integer_var = 1
float_var = 1.0
stri = "hello world"
boole = True
lis = [1, 2, "2"]
menge = set(lis)


print(type(integer_var),
type(float_var),
type(stri),
type(boole),
type(lis)
)


def print_out(itt, stri = "### #### ###"):
    print(stri)
    is_dict = isinstance(itt, dict)
    for item in itt:
        if(is_dict):
            print(f"{item}: {itt[item]}")
        else:
            print(item)

print_out(lis, "--- original list --")

print_out(menge.copy(), "--- Copy Menge ---")

transform = lambda x: int(x) if isinstance(x, str) else x 

print_out(set(map(transform, lis.copy())), "-- gemappte menge --")


# expected: int: 7, string: 3, float: 2, Boolean: 1
bsp_list = ["hallo", 1, 3.2, 4, 1000.0, "h", 24, "m", True, (1,2), {1, 2}]


def summarize_types(liste, tuple_types : tuple):
    result_dict = dict()

    for item in liste:
        item_type = type(item).__name__
        
        if(not isinstance(item, tuple_types)):
            rek_dict = summarize_types(item, tuple_types)
            for k, v in rek_dict.items():
                result_dict[k] = result_dict.get(k, 0) + v
        else:
            result_dict[item_type] = result_dict.get(item_type, 0) + 1
        
    return result_dict


wanted_types = (int, float, str, bool)
print_out(summarize_types(bsp_list, wanted_types ))
```

**Aufgaben:**

    - Eine Textdatei einliest (`data.txt`)

2. Nutze `with open()` korrekt und teste, was passiert, wenn du `try/except` verwendest.
3. Bonus: Lies eine CSV-Datei ohne Pandas ein (`csv`Modul).
    - Zähle, wie viele Zeilen bestimmte Werte in einer Spalte haben.

```python

def trim_and_new_file(file_name : str):
    runner = 0
    with open(file_name, mode='r') as file:
        for line in file:
            runner += 1
            write_line = line.strip().upper()

            try:
                with open(f"data/line_{runner}.txt", mode='x') as newFile:
                    newFile.write(write_line)
            except:
                try:
                    with open(f"data/line_{runner}.txt", mode='w') as newFile:
                        newFile.write(write_line)
                except:
                    print("[ERROR] - No writing possible")
                print(f"File {runner} got overwritten")
            else:
                print("File got created")
    print(f"--- {runner} files have been created ---")


trim_and_new_file("data/data.txt")

import csv
# T = TypeVar('T')
def read_csv_count_values_in_column(filename : str, value : str):
    counter = 0
    dictio = {}
    # anstatt dict() oder {} lieber dictio = defaultdict(list)
    
    first_line = True

    try:
        with open(file=filename, mode="r", newline='') as csv_file:
            csv_reader = csv.DictReader(csv_file)
            for value_csv in csv_reader:

                # not needed: look above
                if first_line:
                    dictio.setdefault
                    dictio = {key: list([val]) for (key, val) in value_csv.items()}
                    first_line = False
                    continue
                # not needed


                dictio.update( {key: list([*dictio[key], val]) for (key, val) in value_csv.items()} )
                

            print(dictio, sep='\n')
            
            res_dict = {k : v.count(value) for (k,v) in dictio.items()}

            print(res_dict)


    except Exception as e:
        print(f"Error - {e}")
    finally:
        print(f"{value} is {counter} - times in the csv-file")

read_csv_count_values_in_column("data/data.csv", 'Berlin')
```
